```r
# Gerando microdados da PNAD Continua pelo metodo automatico

variaveis_selecionadas <- c("V2007",  # Sexo
                            "V2009",  # Idade do morador na data de referência
                            "V2010",  # Cor/raça
                            "V3003A", # Qual é o curso que frequenta
                            "V3009A", # Qual é o curso mais elevado que já frequentou
                            "V3001",  # Sabe ler e escrever
                            "VD4001", # Economicamente ativa/não (Força de trabalho)
                            "VD4002", # Ocupadas/ não ocupadas
                            "VD4020", # Redimentos de trabalho
                            "VD5002", # Renda domiciliar per capita
                            "V4041",  # Código da ocupação1 (função, cargo)
                            "V4010")  # Código da ocupação2 (função, cargo)

#dadosPNADc2016 <- get_pnadc(year=2016, interview=1, labels=TRUE, deflator=TRUE, design=FALSE, vars=variaveis_selecionadas)
#dadosPNADc2017 <- get_pnadc(year=2017, interview=1, labels=TRUE, deflator=TRUE, design=FALSE, vars=variaveis_selecionadas)
#dadosPNADc2018 <- get_pnadc(year=2018, interview=1, labels=TRUE, deflator=TRUE, design=FALSE, vars=variaveis_selecionadas)
dadosPNADc2012 <- get_pnadc(year=2012, interview=1, labels=TRUE, deflator=TRUE, design=FALSE, vars=variaveis_selecionadas)
dadosPNADc2019 <- get_pnadc(year=2019, interview=1, labels=TRUE, deflator=TRUE, design=FALSE, vars=variaveis_selecionadas)

#Unfortunately IBGE didnt release 2020 data yet
dadosPNADc2020 <- get_pnadc(year=2020, interview=1, labels=TRUE, deflator=TRUE, design=FALSE, vars=variaveis_selecionadas)
```

```r
write.csv(dadosPNADc2020, "PNADc2020.csv")
write.csv(dadosPNADc2019, "PNADc2019.csv")
write.csv(dadosPNADc2012, "PNADc2012.csv")
```
