```python
nome = ['Ano',    #Ano
        'UF',     #UF
        'Sexo',   #V2007
        'Idade',  #V2009 
        'Cor',    #V2010
        'Redimentos de trabalho']    #VD4020


df_2019.columns = nome
df_2012.columns = nome
```

```python
dfa19 = df_2019[df_2019['Cor'].str.contains('Amarela')]
dfb19 = df_2019[df_2019['Cor'].str.contains('Branca')]

dfa12 = df_2012[df_2012['Cor'].str.contains('Amarela')]
dfb12 = df_2012[df_2012['Cor'].str.contains('Branca')]
```

```python
def bins(df):

    for i in df['Idade'].index:
    
        if int(df['Idade'][i]) <= 25:
        
            df['Idade'][i] = 25
        
        if df['Idade'][i] > 25 and df['Idade'][i] <= 50:
        
            df['Idade'][i] = 50
        
        if df['Idade'][i] > 50 and df['Idade'][i] <= 70:
        
            df['Idade'][i] = 70
        
        if df['Idade'][i] > 70:
        
            df['Idade'][i] = 71
        
```

```python
bins(dfa19r)
bins(dfa12r)
bins(dfb19r)
bins(dfb12r)
```
